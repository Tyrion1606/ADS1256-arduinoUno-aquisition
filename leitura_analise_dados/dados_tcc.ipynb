# Análise dos dados da leitura cardiaca para o TCC

Importação das bibliotecas usadas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = open('seria_data_heart.csv').read().strip().split(',')
```

### Definição dos parametros da aquisição dos dados

### Carregamento dos dados

```python
x = [float(d) for d in data]    # Dados obtidos
x = x - np.mean(x)
n = len(x)                   # Quantidade de amostras
fs = 1000                       # Frequencia de aquisição dos dados
t_max = n / fs                  # Tempo total de aquisição (s)
t = np.arange(0, t_max, 1/fs)   # Linha do tempo das leituras    
```

### Mostra do gráfico

```python
plt.figure()
plt.plot(t, x)
plt.xlabel('Tempo t (segundos)')
plt.ylabel('x_c(t)')
plt.xlim(2, 4)
plt.show()
```

### FFT

```python
Xdft = np.fft.fft(x)
N = len(Xdft)

plt.figure()
plt.plot(np.arange(N), np.abs(Xdft))
plt.xlabel('Índice de frequência k')
plt.ylabel('|X_{DFT}[k]|')
plt.grid(True)
plt.show()
```

### FT

```python
f_norm = np.linspace(-0.5, 0.5, N)
f = f_norm * fs

f_range = min(fs/2, 50)

plt.figure()
plt.plot(f, np.fft.fftshift(np.abs(Xdft)) / fs)
plt.xlabel('Frequência f (hertz)')
plt.ylabel('|X(f)|')
plt.xlim(-f_range, f_range)
plt.grid(True)
plt.show()
```
